## 2.  Проверка ошибок в данных и их предобработка


### 2.1. Названия, или метки, столбцов датафрейма

Перед началом предобработки необходимо проверить названия столбцов датафрейма.

```python
# Выводим названия всех столбцов датафрейма
games.columns
```

По результатам проверки видно, что названия столбцов содержат пробелы и используют смешанный регистр символов. Такой формат не является ошибкой, однако может затруднять работу с данными и снижать читаемость кода.

Для приведения датафрейма к единому стилю оформления преобразуем названия столбцов в формат `snake_case`: переведём все символы в нижний регистр и заменим пробелы символом подчёркивания.

```python
# Приводим названия столбцов к стилю snake_case
games.columns = games.columns.str.lower().str.replace(' ', '_')
```

```python
# Проверяем результат преобразования
games.columns
```

После преобразования к стилю `snake_case` названия столбцов стали более удобными для дальнейшей работы и обращения к ним в коде.

### 2.2. Типы данных

На этапе первичного знакомства с данными было выявлено, что не все типы данных соответствуют содержимому столбцов.

Особое внимание следует уделить столбцам `eu_sales`, `jp_sales` и `user_score`, которые имеют тип данных `object`, хотя содержат числовые показатели. Наиболее вероятной причиной является наличие строковых значений, пропусков или других некорректных записей, препятствующих автоматическому определению числового типа данных.

Начнём со столбца `user_score`. Проверим, присутствуют ли в столбце значения, которые невозможно преобразовать в число и которые могли привести к некорректному определению типа данных.

```python
# Выводим значения столбца user_score, которые невозможно преобразовать в число
games[pd.to_numeric(games['user_score'], errors='coerce').isna()]['user_score'].unique()
```

Проверка показала, что помимо пропущенных значений в столбце присутствует значение `tbd` (to be determined). Данная запись используется для обозначения отсутствующей оценки и не может быть интерпретирована как число.

Заменим значение `tbd` на пропуск и преобразуем столбец к числовому типу данных.

```python
# Заменяем значение tbd на пропуск
games['user_score'] = games['user_score'].replace('tbd', None)
```

```python
# Преобразуем столбец user_score в числовой тип данных
games['user_score'] = pd.to_numeric(games['user_score'])
```

```python
# Проверяем тип данных столбца user_score
games['user_score'].dtype
```

После замены значения `tbd` и преобразования столбца `user_score` тип данных был успешно изменён на `float64`.

Далее проверим столбцы `eu_sales` и `jp_sales`. Необходимо убедиться, что в данных столбцах отсутствуют нечисловые значения, препятствующие преобразованию к числовому типу данных.

```python
# Выводим значения столбца eu_sales, которые невозможно преобразовать в число
games.loc[pd.to_numeric(games['eu_sales'], errors='coerce').isna(),
    'eu_sales'].unique()
```

```python
# Выводим значения столбца eu_sales, которые невозможно преобразовать в число
games.loc[pd.to_numeric(games['jp_sales'], errors='coerce').isna(),
    'jp_sales'].unique()
```

Проверка показала, что в столбцах `eu_sales` и `jp_sales` присутствует значение `unknown`, которое не может быть интерпретировано как число.

Вероятно, данная запись используется для обозначения отсутствующей информации о продажах в соответствующем регионе. Заменим значение `unknown` на пропуск и преобразуем оба столбца к числовому типу данных.

```python
# Заменяем значение unknown на пропуск
games['eu_sales'] = games['eu_sales'].replace('unknown', None)
games['jp_sales'] = games['jp_sales'].replace('unknown', None)
```

```python
# Преобразуем столбцы продаж к числовому типу данных
games['eu_sales'] = pd.to_numeric(games['eu_sales'])
games['jp_sales'] = pd.to_numeric(games['jp_sales'])
```

```python
# Проверяем типы данных после преобразования
games[['eu_sales', 'jp_sales']].dtypes
```

Таким образом, все столбцы, в которых были обнаружены некорректные типы данных, успешно преобразованы.

Столбцы `user_score`, `eu_sales` и `jp_sales` теперь имеют числовой тип данных, что позволит корректно выполнять расчёты и статистический анализ

Поскольку количество таких записей составляет всего 2 строки (около 0,01% от общего объёма данных), их удаление не окажет существенного влияния на результаты исследования. Поэтому удалим данные записи из датафрейма.

```python
# Удаляем строки с пропусками в названии игры и жанре
games = games.dropna(subset=['name', 'genre'])
```

```python
# Проверяем результат удаления
games[games['name'].isna() | games['genre'].isna()]
```

Далее рассмотрим пропуски в столбце `year_of_release`.

Данный столбец играет важную роль в исследовании, поскольку на его основе будет формироваться выборка игр за период с 2000 по 2013 год. Поэтому необходимо определить возможность восстановления пропущенных значений или выбрать другой способ их обработки.

```python
# Выводим строки с пропусками в году выпуска
games[games['year_of_release'].isna()]
```

При просмотре строк с пропущенным годом выпуска видно, что часть игр содержит другую полезную информацию: название, платформу, жанр, продажи, оценки и рейтинг.

Однако восстановить год выпуска автоматически по имеющимся данным нельзя. Название игры не всегда однозначно указывает на год выхода, а одна и та же игра могла выпускаться на разных платформах в разные годы.

Так как столбец `year_of_release` необходим для дальнейшей фильтрации данных по периоду с 2000 по 2013 год, строки без года выпуска не подходят для дальнейшего анализа. Доля таких строк составляет 1,62%, поэтому их удаление не должно существенно повлиять на результат исследования.

```python
# Удаляем строки с пропусками в году выпуска
games = games.dropna(subset=['year_of_release'])
```

```python
# Преобразуем столбец year_of_release в целочисленный тип данных
games['year_of_release'] = games['year_of_release'].astype('int')
```

```python
# Проверяем тип данных столбца year_of_release
games['year_of_release'].dtype
```

```python
# Проверяем, остались ли пропуски в ключевых столбцах
games[['name', 'genre', 'year_of_release']].isna().sum()
```

Рассмотрим пропуски в столбцах `critic_score`, `user_score` и `rating`.

На данные столбцы приходится наибольшее количество пропущенных значений в датафрейме. Все они содержат информацию об оценках игры или её возрастном рейтинге.

Поскольку отсутствие оценки или рейтинга само по себе не является ошибкой в данных, важно определить, целесообразно ли заполнять такие пропуски искусственными значениями или лучше сохранить их в исходном виде.

```python
# Выводим количество пропусков в столбцах с оценками и рейтингом
games[['critic_score', 'user_score', 'rating']].isna().sum()
```

Пропуски в столбцах `critic_score` и `user_score`, вероятно, связаны с отсутствием достаточного количества обзоров критиков или пользовательских оценок для отдельных игр.

Заполнять данные пропуски средними или медианными значениями нецелесообразно, поскольку это может привести к искажению результатов анализа. Отсутствие оценки не означает, что игра имеет среднюю или какую-либо другую оценку.

Поэтому пропуски в столбцах `critic_score` и `user_score` будут сохранены без изменений.

Рассмотрим отдельно столбец `rating`. Поскольку данный признак является категориальным, вместо удаления большого количества записей можно использовать специальное значение-индикатор. Предварительно проверим, отсутствует ли выбранное значение среди существующих категорий рейтинга ESRB.

```python
# Выводим уникальные значения столбца rating
games['rating'].dropna().unique()
```

```python
# Заменяем пропуски в рейтинге на значение-индикатор
games['rating'] = games['rating'].fillna('UNKNOWN')
```

```python
# Проверяем количество пропусков после замены
games['rating'].isna().sum()
```

Проверка показала, что значение `UNKNOWN` отсутствует среди существующих категорий рейтинга ESRB, поэтому его можно использовать в качестве значения-индикатора.

Пропуски в столбце `rating` были заменены на значение `UNKNOWN`, которое обозначает отсутствие информации о возрастном рейтинге игры и не пересекается с существующими категориями ESRB.

Таким образом, все пропуски в столбце `rating` были обработаны, а пропуски в столбцах `critic_score` и `user_score` сохранены без изменений.

### 2.4. Явные и неявные дубликаты в данных

После обработки пропущенных значений необходимо проверить данные на наличие явных и неявных дубликатов.

Неявные дубликаты могут возникать из-за различий в регистре символов, опечаток или различных способов написания одних и тех же значений. Особое внимание следует уделить категориальным столбцам, таким как название платформы, жанр игры и возрастной рейтинг.

После проверки и устранения возможных неявных дубликатов необходимо проверить датафрейм на наличие полностью совпадающих строк.

И при необходимости провести нормализацию данных с текстовыми значениями. Названия жанров игр привести к нижнему регистру, а названия рейтинга - к верхнему.

```python
# Выводим уникальные значения платформ
sorted(games['platform'].unique())
```

```python
# Выводим уникальные значения жанров
sorted(games['genre'].unique())
```

```python
# Выводим уникальные значения рейтингов
sorted(games['rating'].unique())
```

Проверка уникальных значений показала, что в столбце `platform` неявные дубликаты отсутствуют. Названия платформ записаны единообразно.

В столбце `genre` были обнаружены неявные дубликаты, связанные с различием регистра символов. Например, один и тот же жанр представлен значениями `ACTION` и `Action`, `SPORTS` и `Sports`, `RACING` и `Racing`.

Для устранения подобных дубликатов приведём все значения столбца `genre` к единому формату записи.

Также при анализе уникальных значений столбца `rating` было обнаружено, что категории `K-A` и `E` фактически обозначают одну и ту же возрастную группу.

Рейтинг `K-A` (Kids to Adults) использовался организацией ESRB до 1998 года и впоследствии был заменён на рейтинг `E` (Everyone). Поскольку данные категории являются эквивалентными, для обеспечения единообразия данных заменим значение `K-A` на `E`.

```python
# Приводим названия жанров к единому регистру
games['genre'] = games['genre'].str.lower()
```

```python
# Проверяем результат преобразования
sorted(games['genre'].unique())
```

```python
# Заменяем устаревший рейтинг K-A на E
games['rating'] = games['rating'].replace('K-A', 'E')
```

```python
# Проверяем уникальные значения рейтингов
sorted(games['rating'].unique())
```

## 4. Категоризация данных

После формирования необходимого временного среза выполним категоризацию игр по пользовательским и экспертным оценкам.

Пользовательские оценки будут разделены на следующие категории:

- низкая оценка - от 0 до 3, не включая правую границу интервала
- средняя оценка - от 3 до 8, не включая правую границу интервала
- высокая оценка - от 8 до 10 включительно

Оценки критиков будут разделены аналогичным образом:

- низкая оценка - от 0 до 30, не включая правую границу интервала
- средняя оценка - от 30 до 80, не включая правую границу интервала
- высокая оценка - от 80 до 100 включительно

Для удобства анализа будут созданы отдельные столбцы с категориями пользовательских и экспертных оценок.
Кроме того, будет определён топ-7 платформ, на которых за рассматриваемый период было выпущено наибольшее количество игр.

Начнём с пользовательских оценок. Для каждой игры определим категорию оценки в соответствии с указанными интервалами.

```python
# Создаем функцию категоризации пользовательских оценок
def user_score_category(score):
    if pd.isna(score):
        return 'нет оценки'
    elif 0 <= score < 3:
        return 'низкая'
    elif 3 <= score < 8:
        return 'средняя'
    elif 8 <= score <= 10:
        return 'высокая'
```

```python
# Создаем новый столбец с категориями пользовательских оценок
df_actual['user_score_category'] = df_actual['user_score'].apply(user_score_category)
```

```python
# Подсчитываем количество игр в каждой категории пользовательских оценок
df_actual['user_score_category'].value_counts()
```

Пользовательские оценки разделены на категории в соответствии с условиями задания.

Наибольшее количество игр относится к категории без пользовательской оценки - 5195 записей. Это подтверждает вывод, сделанный ранее при анализе пропусков: значительная часть игр не имеет пользовательских оценок.

Среди игр, для которых оценки доступны, преобладает категория со средней оценкой - 4820 игр. Высокую оценку получили 2621 игра, тогда как низкая оценка встречается значительно реже - всего у 145 игр.

Таким образом, большинство игр с доступными пользовательскими оценками имеют средний или высокий уровень оценки.

Аналогичным образом выполним категоризацию оценок критиков.
Для определения категорий будут использоваться интервалы, указанные ранее.

```python
# Создаем функцию категоризации оценок критиков
def critic_score_category(score):
    if pd.isna(score):
        return 'нет оценки'
    elif 0 <= score < 30:
        return 'низкая'
    elif 30 <= score < 80:
        return 'средняя'
    elif 80 <= score <= 100:
        return 'высокая'
```

```python
# Создаем новый столбец с категориями оценок критиков
df_actual['critic_score_category'] = df_actual['critic_score'].apply(critic_score_category)
```

```python
# Подсчитываем количество игр в каждой категории оценок критиков
df_actual['critic_score_category'].value_counts()
```

Оценки критиков также успешно разделены на категории в соответствии с заданными интервалами.

Наибольшее количество игр относится к категории без оценки критиков - 5612 записей. Это соответствует результатам анализа пропусков, выполненного на этапе предобработки данных.

Среди игр с доступными оценками критиков преобладает категория со средней оценкой - 5422 игры. Высокую оценку получили 1692 игры, тогда как низкие оценки встречаются достаточно редко - всего у 55 игр.

Таким образом, как и в случае с пользовательскими оценками, большинство игр с доступными оценками критиков относятся к категории средних или высоких оценок.

Завершающим этапом категоризации выделим платформы, на которых за рассматриваемый период было выпущено наибольшее количество игр.

Для этого сгруппируем данные по платформам и определим семь платформ с максимальным количеством выпущенных игр.

```python
# Определяем топ-7 платформ по количеству выпущенных игр
top_platforms = (df_actual['platform'].value_counts().head(7))
top_platforms
```

По количеству выпущенных игр за период с 2000 по 2013 год лидирует платформа `PS2`, для которой было выпущено 2127 игр. Практически такой же результат показывает платформа `DS` - 2120 игр.
